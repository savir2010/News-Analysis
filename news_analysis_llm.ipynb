```python
from langchain.document_loaders import TextLoader
import os
```

```python
from langchain.document_loaders import UnstructuredURLLoader
```

```python
loaders = UnstructuredURLLoader(urls=[
    "https://www.moneycontrol.com/news/technology/meesho-churn-continues-as-cxo-harsh-chaudhary-departs-to-join-rival-flipkart-12483281.html",
    "https://www.moneycontrol.com/news/technology/brazenly-wrong-fiitjee-founders-salary-is-not-a-right-email-sparks-concern-from-hr-legal-experts-12456371.html"])
data = loaders.load()
len(data)
```

```python
data[0].metadata
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

docs = text_splitter.split_documents(data)
```

```python
docs = text_splitter.split_documents(data)
```

```python
docs[0]
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.llms import GooglePalm
```

```python
api_key = 'API KEY'
```

```python
llm = GooglePalm(google_api_key=api_key,temperature=0)
```

```python
from langchain.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
vectordb = FAISS.from_documents(documents=docs,embedding=embeddings)
```

```python
retriever = vectordb.as_retriever()
```

```python
retriever.get_relevant_documents("Who is  FIITJEE's Founder and Managing Director")
```

```python
prompt_template = """Given the following article and a question, generate an answer based on this article only.
If the answer is not found in the article, kindly state "Answer is not stated in the article" Do not make anything up.

CONTEXT: {context}

QUESTION: {question}"""

PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
    )
```

```python
chain = RetrievalQA.from_chain_type(llm=llm,
                                    chain_type="stuff",
                                    retriever=retriever,
                                    input_key="query",
                                    return_source_documents=True,
                                    chain_type_kwargs={"prompt": PROMPT})
```

```python
chain("Who is  FIITJEE's Founder and Managing Director")
```
